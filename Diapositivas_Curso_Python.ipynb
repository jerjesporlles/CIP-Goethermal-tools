```python
# Crear funciones en Python
def calculadora ( variable1, variable2): # Inputs
    respuesta_1 = variable1*2+variable2/50+variable1*variable2 - variable2**3
    return respuesta_1 # Outputs
```

```python
r = calculadora(1,2)
print("El resultado es : " , r)
```

```python
# Crear funciones en Python
def calculadora_2 ( variable1, variable2):
    respuesta_1 = variable1*2+variable2/50+variable1*variable2 - variable2**3
    respuesta_2 = variable1+variable2
    return respuesta_1, respuesta_2
```

```python
s,t = calculadora_2(1,2)
print("El resultado 1 es : " , s)
print("El resultado 2 es : " , t)
```

```python
df_clean = df.dropna(subset=['NPHI','DT','RHOB','GR','SP','RT'],axis=0, how='any')
df_clean
```

```python
limite_inf=10000
limite_sup=10050
df_limites=df_clean[limite_inf:limite_sup]
df_limites
```

```python
df_limites.columns
```

```python
seleccionar_columnas = ['NPHI','DT','RHOB','GR','SP','RT']
df_filtrado = df_limites[seleccionar_columnas ]
df_filtrado
```

```python
fig1, ax = plt.subplots( figsize=(20,4))
ax.plot(df_filtrado.index,df_filtrado['NPHI'], color = 'green')
ax.invert_yaxis()
ax.set_xlabel('NPHI')
ax.set_ylabel("Depth(ft)")
ax.grid()
```

```python
#PLoteado
def plotter():
 f, ax = plt.subplots(nrows=1, ncols=6, figsize=(16,10))
 logs = ['NPHI','DT','RHOB','GR','SP','RT']
 colors = ['green','orange','blue','red','black','brown']
 for i,log,color in zip(range(6),logs,colors):
   ax[i].plot(df_filtrado[log], df_filtrado.index, color=color)
   ax[i].invert_yaxis()
   ax[i].set_xlabel(log)
   ax[i].set_ylabel("Depth(ft)")
   ax[i].grid()
#plotter()
```
